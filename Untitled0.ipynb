```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def draw_graph(G):
  nx.draw_networkx(G, pos=nx.kamada_kawai_layout(G), \
                   arrows=True if nx.is_directed(G) else None,\
                   node_size=500, node_color="yellow", font_size=20)
```

```python
random_graph = nx.erdos_renyi_graph(n=10, p=0.3)
draw_graph(random_graph)
```

```python
def checkNodes(G):
  print("Set of vertices: {0}".format(list(G.nodes())))
```

```python
checkNodes(random_graph)
```

```python
def checkEdges(G):
  print("Set of edges: {0}".format(list(G.edges())))
```

```python
checkEdges(random_graph)
```

```python
def checkDegrees(G, nodes_to_check_degree):
  for node in nodes_to_check_degree:
    if node in G.nodes():
      print("Degree of {0}:{1}".format(node, G.degree(node)))
    else:
      print("Node {0} is not in Graph".format(node))
```

```python
checkDegrees(random_graph, [0,1,2])
```

```python
def checkDegreeSequence(G):
  print("Degree seqeuence:{0}".format(sorted([G.degree(n) for n in G.nodes()])))

checkDegreeSequence(random_graph)
```

```python
def plot_degree_dist(G):
  degrees = [G.degree(n) for n in G.nodes()]
  plt.hist(degrees)
  plt.show()
```

```python
er = nx.erdos_renyi_graph(n = 100, p=0.3)
nx.draw(er)
nx.algorithms.distance_measures.diameter(er)
```

```python

plot_degree_dist(er)
nx.algorithms.cluster.average_clustering(er)
```
