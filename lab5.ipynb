Training RMSE: 0.12768967421762192
Testing RMSE: 0.1458346449094924

```python
import numpy as np
from math import sqrt




#utility function to return data in array from given file
def loadsample(path):
    return np.loadtxt(fname = path, dtype = np.float64, delimiter = '\t', skiprows = 1)
    
    
    
#problem 1: baseline linear regression
def problem1(samples):
    
    
    #import training data
    train = loadsample('crime-train.txt')
    
    
    #split into features and result
    trainsplit = np.split(train, [1], 1)
    b = np.full((len(trainsplit[0]), 1), 1)
    trainX = np.append(trainsplit[1], b, axis = 1)
    trainY = trainsplit[0]
    
    
    #calculate weights (broken down for readability)
    trainXt = np.transpose(trainX)
    xtxprod = np.matmul(trainXt, trainX)
    invm = np.linalg.inv(xtxprod)
    invmxt = np.matmul(invm, trainXt)
    w = np.matmul(invmxt, trainY)
    
    #generate predictions
    return (np.matmul(samples, w))
    
    

          
          
          
          
train = loadsample('crime-train.txt')
trainsplit = np.split(train, [1], 1)
b = np.full((len(trainsplit[0]), 1), 1)
trainX = np.append(trainsplit[1], b, axis = 1)
trainY = trainsplit[0]

rmse = 0
pred = problem1(trainX)
print(pred)
for i in range(len(trainX)):
    rmse += (pred[i][0] - trainY[i][0]) ** 2
    
rmse /= len(trainX)
print("\nTraining RMSE: " + str(sqrt(rmse)))






test = loadsample('crime-test.txt')
testsplit = np.split(test, [1], 1)
b = np.full((len(testsplit[0]), 1), 1)
testX = np.append(testsplit[1], b, axis = 1)
testY = testsplit[0]
    
rmse = 0
pred = problem1(testX)
for i in range(len(testX)):
    rmse += (pred[i][0] - testY[i][0]) ** 2
    
rmse /= len(testX)
print("\nTesting RMSE: " + str(sqrt(rmse)))
    
```

```python
import numpy as np
from math import sqrt




#utility function to return data in array from given file
def loadsample(path):
    return np.loadtxt(fname = path, dtype = np.float64, delimiter = '\t', skiprows = 1)
    
    
    
#problem 1: baseline linear regression
def problem2(samples):
    
    #import training data
    train = loadsample('crime-train.txt')
    
    
    #split into features and result
    trainsplit = np.split(train, [1], 1)
    b = np.full((len(trainsplit[0]), 1), 1)
    trainX = np.append(trainsplit[1], b, axis = 1)
    trainY = trainsplit[0]
    
    
    #calculate weights (broken down for readability)
    diag = np.zeros((len(trainX[0]), len(trainX[0])), dtype = np.float64)
    for i in range(len(trainX[0])):
        diag[i][i] = 100
    
    trainXt = np.transpose(trainX)
    xtxprod = np.matmul(trainXt, trainX) + diag
    invm = np.linalg.inv(xtxprod)
    invmxt = np.matmul(invm, trainXt)
    w = np.matmul(invmxt, trainY)
    
    
    
    #generate predictions
    return (np.matmul(samples, w))
    

          
          
          
          
train = loadsample('crime-train.txt')
trainsplit = np.split(train, [1], 1)
b = np.full((len(trainsplit[0]), 1), 1)
trainX = np.append(trainsplit[1], b, axis = 1)
trainY = trainsplit[0]
    
rmse = 0
pred = problem2(trainX)
print(pred)
for i in range(len(trainX)):
    rmse += (pred[i][0] - trainY[i][0]) ** 2
    
rmse /= len(trainX)
print("\nTraining RMSE: " + str(sqrt(rmse)))






test = loadsample('crime-test.txt')
testsplit = np.split(test, [1], 1)
b = np.full((len(testsplit[0]), 1), 1)
testX = np.append(testsplit[1], b, axis = 1)
testY = testsplit[0]
    
rmse = 0
pred = problem2(testX)
for i in range(len(testX)):
    rmse += (pred[i][0] - testY[i][0]) ** 2
    
rmse /= len(testX)
print("\nTesting RMSE: " + str(sqrt(rmse)))
    
```
